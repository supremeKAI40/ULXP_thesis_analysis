```python
# Organize your count arrays for easy iteration
count_arrays = [count227, count233, count244, count251, count255, count261, count267, count281]

# Create a figure with a 4x2 grid of subplots
fig, ax = plt.subplots(4, 2, figsize=(30, 20))  # Adjust figsize for better visibility

# Flatten the axes array for easier indexing
ax = ax.flatten()

# Plot each count array against phase227
for i, count in enumerate(count_arrays):
    ax[i].plot(phase227, count)  # Dynamic label based on count number
    ax[i].set_title(f'Count {227 + 6 * i} vs Phase 227', fontsize=14)  # Dynamic title
    ax[i].set_xlabel('Phase 227', fontsize=14)
    ax[i].set_ylabel('Count', fontsize=14)
    ax[i].grid()
    ax[i].tick_params(axis='both', labelsize=14)

# Adjust layout for clarity
plt.tight_layout()
plt.show()
```
