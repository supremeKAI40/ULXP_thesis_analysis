```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
from numpy.lib.recfunctions import append_fields
```

```python
search_directory="./reduced_output"
obsid_pattern="6050*"

search_pattern_new=os.path.join(search_directory,obsid_pattern,"*night_barycorr_orbit_piexpiex.evt")
search_pattern_old=os.path.join(search_directory,obsid_pattern,"*cl_night_barycorr.evt")
matching_files_new = glob.glob(search_pattern_new)
matching_files_old=glob.glob(search_pattern_old)
```

```python
len(matching_files_new)
```

```python
# Function to extract ObsID from the file path
def extract_obsid(file_path):
    # Assuming the ObsID is the directory name after `search_directory`
    parts = os.path.normpath(file_path).split(os.sep)
    return parts[1]  # Adjust index based on your path structure

# Create dictionaries mapping ObsIDs to file paths
obsid_to_files_new = {extract_obsid(path): path for path in matching_files_new}
obsid_to_files_old = {extract_obsid(path): path for path in matching_files_old}

# Find common ObsIDs
common_obsids = set(obsid_to_files_new.keys()) & set(obsid_to_files_old.keys())

# Filter the lists to include only paths with matching ObsIDs
filtered_matching_files_new = [obsid_to_files_new[obsid] for obsid in common_obsids]
filtered_matching_files_old = [obsid_to_files_old[obsid] for obsid in common_obsids]
```

```python
def extract_obsid(filename):
    # Assuming the obsID is a 4-digit number at a specific location in the file name (e.g., '6050xxxx')
    return filename.split('/')[-1][2:12]  # Adjust the indexing as needed


for new_file, old_file in zip(filtered_matching_files_new, filtered_matching_files_old):
    print(new_file)

    # Extract obsID from both new and old file names
    obsid_new = extract_obsid(new_file)
    obsid_old = extract_obsid(old_file)


    # Check if obsIDs match
    if obsid_new != obsid_old:
        print(f"Warning: obsID mismatch - {obsid_new} vs {obsid_old}")
        continue  # Skip this pair of files if the obsIDs do not match
    

    hdul_new=fits.open(new_file)
    time_new=hdul_new[1].data['TIME'].astype(np.float128)

    print(old_file)
    hdul_old=fits.open(old_file)

    hdr_old=hdul_old[1].header
    data_old = hdul_old[1].data  # Accessing the table data
    
    new_column = fits.Column(name="ORBITTIME", format="1D", unit="seconds", array=time_new)
    updated_columns = fits.ColDefs(hdul_old[1].columns + new_column)
    
    hdul_old[1].columns.add_col(new_column)

    # updated_table_hdu = fits.BinTableHDU.from_columns(updated_columns, header=hdr_old)

    # hdul_old[1]=updated_table_hdu

    output_path=new_file.replace(".evt","_yes.evt")
    
    print("Saved final files as: ",output_path)
    
    hdul_old.writeto(output_path,overwrite=True)
```
