```python
from astropy.io import fits
import matplotlib.pyplot as plt
```

```python
flux=fits.open('phase_resolved_flux_with_errors.fits')
flux[1].header

rms=fits.open('rms_deviation_flux_errors.fits')
rms[1].header
```

```python
plt.figure(figsize=(14,5))
plt.errorbar(flux[1].data['PHASE'],flux[1].data['avg_flux'],yerr=flux[1].data['avg_flux_error']
             ,elinewidth=1,fmt='-o',capsize=1,color='r')

plt.errorbar(rms[1].data['PHASE'],rms[1].data['RMS'],yerr=rms[1].data['RMS_ERROR']
             ,elinewidth=1,fmt='-o',capsize=1,color='b')
plt.grid()
plt.title('Folded light curve for ObsID 6050390261')
plt.xlabel('Phase(0-2)')
plt.ylabel('Count rate')
```

```python
plt.figure(figsize=(14,5))

plt.errorbar(rms[1].data['PHASE'],rms[1].data['RMS']/flux[1].data['avg_flux'],
             elinewidth=1,fmt='-o',capsize=1,color='b')
plt.grid()
plt.title('Ratio of RMS to Avg count rate for ObsID 6050390261')
plt.xlabel('Phase(0-2)')
plt.ylabel('RMS/Count rate')
```
