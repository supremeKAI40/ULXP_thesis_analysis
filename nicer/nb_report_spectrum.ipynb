```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))  # Set the figure size for better clarity

# Scatter plot with customizations
plt.scatter(df['MJD'] , df['Lumin'], color='b', alpha=0.7, label='Luminosity')

# Adding labels and title
plt.xlabel('MJD ', fontsize=12)
plt.ylabel('Luminosity', fontsize=12)
plt.title('Scatter Plot of Luminosity vs. MJD (Adjusted)', fontsize=14)

# Adding grid and setting axis limits if needed
plt.grid(True)
plt.tight_layout()

# Optionally, set the x and y axis limits based on the data range
# plt.xlim(min(df['MJD'] - 60000), max(df['MJD'] - 60000))
# plt.ylim(min(df['Lumin']), max(df['Lumin']))

# Show plot with legend
plt.legend()
plt.show()
```

```python
def clean_nh_value(value):
    try:
        # Try converting to float
        return float(value)
    except ValueError:
        # Handle 'frozen' or other non-numeric cases
        return np.nan
df['nH'] = df['nH'].apply(clean_nh_value)

yerr = [np.abs(df['nh_low']), np.abs(df['nh_hi'])]
```

```python
# Create two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5), constrained_layout=True)

# First plot: Absorption column density vs Luminosity
axes[0].errorbar(df['Lumin'], df['nH'], yerr=yerr, fmt='o', capsize=5, color='b', label='nH values')
axes[0].set_title("Absorption Column Density vs Luminosity", fontsize=12)
axes[0].set_xlabel("Luminosity (erg/s)", fontsize=11)
axes[0].set_ylabel(r"$n_H$ ($\times 10^{22}$ atoms/cm$^2$)", fontsize=11)
axes[0].grid(True, linestyle='--', alpha=0.6)
axes[0].legend()

# Second plot: Absorption column density vs MJD
axes[1].errorbar(df['MJD'], df['nH'], yerr=yerr, fmt='o', capsize=5, color='r', label='nH values')
axes[1].set_title("Absorption Column Density vs MJD", fontsize=12)
axes[1].set_xlabel("Modified Julian Date (MJD)", fontsize=11)
axes[1].set_ylabel(r"$n_H$ ($\times 10^{22}$ atoms/cm$^2$)", fontsize=11)
axes[1].grid(True, linestyle='--', alpha=0.6)
axes[1].legend()

# Show the plots
plt.show()
```

```python
# Assuming yerr is already defined as:
yerr = [np.abs(df['pi_low']), np.abs(df['pi_hi'])]

# Create two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5), constrained_layout=True)

# First plot: Powerlaw photon index vs Luminosity
axes[0].errorbar(df['Lumin'], df['PI'], yerr=yerr, fmt='o', capsize=5, color='b', label='PI values')
axes[0].set_title("Powerlaw Photon Index vs Luminosity", fontsize=12)
axes[0].set_xlabel("Luminosity (erg/s)", fontsize=11)
axes[0].set_ylabel("Photon Index (PI)", fontsize=11)
axes[0].grid(True, linestyle='--', alpha=0.6)
axes[0].legend()

# Second plot: Powerlaw photon index vs MJD
axes[1].errorbar(df['MJD'], df['PI'], yerr=yerr, fmt='o', capsize=5, color='r', label='PI values')
axes[1].set_title("Powerlaw Photon Index vs MJD", fontsize=12)
axes[1].set_xlabel("Modified Julian Date (MJD)", fontsize=11)
axes[1].set_ylabel("Photon Index (PI)", fontsize=11)
axes[1].grid(True, linestyle='--', alpha=0.6)
axes[1].legend()

# Show the plots
plt.show()
```

```python
# Clean the data for lineE, line_low, line_hi, lineSigma, sigma_low, sigma_hi
def clean_line_value(value):
    try:
        return float(value)
    except ValueError:
        return np.nan

# Apply the clean_line_value function to kT, kt_low, and kt_hi
df['kT'] = df['kT'].apply(clean_line_value)
df['kt_low'] = df['kt_low'].apply(clean_line_value)
df['kt_hi'] = df['kt_hi'].apply(clean_line_value)

# Filter out rows where kT is 0
df_filtered = df[df['kT'] > 0]

# Define error bars using kt_low and kt_hi
yerr = [np.abs(df_filtered['kt_low']), np.abs(df_filtered['kt_hi'])]

# Create two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5), constrained_layout=True)

# First plot: Disk temperature vs Luminosity
axes[0].errorbar(df_filtered['Lumin'], df_filtered['kT'], yerr=yerr, fmt='o', capsize=5, color='b', label='kT values')
axes[0].set_title("Disk Temperature vs Luminosity", fontsize=12)
axes[0].set_xlabel("Luminosity (erg/s)", fontsize=11)
axes[0].set_ylabel("kT (keV)", fontsize=11)
axes[0].grid(True, linestyle='--', alpha=0.6)
axes[0].legend()

# Second plot: Disk temperature vs MJD
axes[1].errorbar(df_filtered['MJD'], df_filtered['kT'], yerr=yerr, fmt='o', capsize=5, color='r', label='kT values')
axes[1].set_title("Disk Temperature vs MJD", fontsize=12)
axes[1].set_xlabel("Modified Julian Date (MJD)", fontsize=11)
axes[1].set_ylabel("kT (keV)", fontsize=11)
axes[1].grid(True, linestyle='--', alpha=0.6)
axes[1].legend()

# Show the plots
plt.show()
```

```python


df['lineE'] = df['lineE'].apply(clean_line_value)
df['line_low'] = df['line_low'].apply(clean_line_value)
df['line_hi'] = df['line_hi'].apply(clean_line_value)

df['lineSigma'] = df['lineSigma'].apply(clean_line_value)
df['sigma_low'] = df['sigma_low'].apply(clean_line_value)
df['sigma_hi'] = df['sigma_hi'].apply(clean_line_value)

# Define error bars for lineE and lineSigma
yerr_lineE = [np.abs(df['line_low'][df['lineE'] > 0]), np.abs(df['line_hi'][df['lineE'] > 0])]
yerr_lineSigma = [np.abs(df['sigma_low'][df['lineE'] > 0]), np.abs(df['sigma_hi'][df['lineE'] > 0])]

# Create a 2x2 subplot grid
fig, axes = plt.subplots(2, 2, figsize=(12, 10), constrained_layout=True)

# First plot: Gaussian line energy vs Luminosity
axes[0, 0].errorbar(df['Lumin'][df['lineE'] > 0], df['lineE'][df['lineE'] > 0], yerr=yerr_lineE, fmt='o', capsize=5, color='teal', label='Line Energy values')
axes[0, 0].set_title("Gaussian Line Energy vs Luminosity", fontsize=12)
axes[0, 0].set_xlabel("Luminosity (erg/s)", fontsize=11)
axes[0, 0].set_ylabel("Line Energy (keV)", fontsize=11)
axes[0, 0].grid(True, linestyle='--', alpha=0.6)
axes[0, 0].legend()

# Second plot: Gaussian line energy vs MJD
axes[0, 1].errorbar(df['MJD'][df['lineE'] > 0], df['lineE'][df['lineE'] > 0], yerr=yerr_lineE, fmt='o', capsize=5, color='darkcyan', label='Line Energy values')
axes[0, 1].set_title("Gaussian Line Energy vs MJD", fontsize=12)
axes[0, 1].set_xlabel("Modified Julian Date (MJD)", fontsize=11)
axes[0, 1].set_ylabel("Line Energy (keV)", fontsize=11)
axes[0, 1].grid(True, linestyle='--', alpha=0.6)
axes[0, 1].legend()

# Third plot: Gaussian line width vs Luminosity
axes[1, 0].errorbar(df['Lumin'][df['lineE'] > 0], df['lineSigma'][df['lineE'] > 0], yerr=yerr_lineSigma, fmt='o', capsize=5, color='teal', label='Line Sigma values')
axes[1, 0].set_title("Gaussian Line Width vs Luminosity", fontsize=12)
axes[1, 0].set_xlabel("Luminosity (erg/s)", fontsize=11)
axes[1, 0].set_ylabel("Line Width (keV)", fontsize=11)
axes[1, 0].grid(True, linestyle='--', alpha=0.6)
axes[1, 0].legend()

# Fourth plot: Gaussian line width vs MJD
axes[1, 1].errorbar(df['MJD'][df['lineE'] > 0], df['lineSigma'][df['lineE'] > 0], yerr=yerr_lineSigma, fmt='o', capsize=5, color='darkcyan', label='Line Sigma values')
axes[1, 1].set_title("Gaussian Line Width vs MJD", fontsize=12)
axes[1, 1].set_xlabel("Modified Julian Date (MJD)", fontsize=11)
axes[1, 1].set_ylabel("Line Width (keV)", fontsize=11)
axes[1, 1].grid(True, linestyle='--', alpha=0.6)
axes[1, 1].legend()

# Show the plots
plt.show()
```

```python
# Assuming the reduced/chi-sq values are in a column called 'reduced_chi_sq'
def extract_chi_squared(value):
    try:
        chi_sq, dof = value.split('/')
        return float(chi_sq), int(dof)
    except ValueError:
        return np.nan, np.nan

# Apply the extraction function to split 'reduced/chi-sq' values
df[['chi_sq', 'dof']] = df['reduced/chi-sq'].apply(extract_chi_squared).apply(pd.Series)

# Calculate reduced chi-squared
df['reduced_chi_sq'] = df['chi_sq'] / df['dof']

# Plotting reduced chi-squared vs Luminosity
plt.errorbar(df['Lumin'], df['reduced_chi_sq'], fmt='o', capsize=5, color='purple', label='Reduced Chi-squared vs Luminosity')
plt.title("Reduced Chi-squared vs Luminosity")
plt.xlabel("Luminosity (erg/s)")
plt.ylabel("Reduced Chi-squared")
plt.grid(True)
plt.legend()
plt.show()

# Plotting reduced chi-squared vs MJD
plt.errorbar(df['MJD'], df['reduced_chi_sq'], fmt='o', capsize=5, color='purple', label='Reduced Chi-squared vs MJD')
plt.title("Reduced Chi-squared vs MJD")
plt.xlabel("Modified Julian Date (MJD)")
plt.ylabel("Reduced Chi-squared")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Mask to check if kT is '0' (we treat '0' as missing in this case)
mask_kT_exists = df['kT'] == 0  # True where kT is 0, False where kT is non-zero

# Set colors: 'red' where kT is 0, 'blue' where kT is non-zero
colors = mask_kT_exists.map({True: 'red', False: 'blue'})

# Define error bars using PI_low and PI_hi
yerr = [np.abs(df['pi_low']), np.abs(df['pi_hi'])]

# Create scatter plot with error bars
plt.errorbar(df['Lumin'], df['PI'], yerr=yerr, fmt='o', capsize=5, color='gray', alpha=0.2, label='Error bars')
plt.scatter(df['Lumin'], df['PI'], c=colors, alpha=0.7)

# Add title and labels
plt.title("Photon Index vs Luminosity (Color coded by kT existence)")
plt.xlabel("Luminosity (erg/s)")
plt.ylabel("Photon Index (PI)")

# Add a custom legend
plt.scatter([], [], color='blue', edgecolors='black', label='kT exists (non-zero)', alpha=0.7)
plt.scatter([], [], color='red', edgecolors='black', label='kT not exists (zero)', alpha=0.7)

# Display the legend
plt.legend()

# Show grid and plot
plt.grid(True)
plt.show()

```
