```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("reporting_spectra.csv")
df.columns
```

```python
def clean_nh_value(value):
    try:
        # Try converting to float
        return float(value)
    except ValueError:
        # Handle 'frozen' or other non-numeric cases
        return np.nan
df['nH'] = df['nH'].apply(clean_nh_value)

yerr = [np.abs(df['nh_low']), np.abs(df['nh_hi'])]
```

```python
yerr = [np.abs(df['nh_low']), np.abs(df['nh_hi'])]
plt.errorbar(df['Lumin'],df['nH'],yerr=yerr,fmt='o', capsize=5, label='nH values')
plt.title("absorption column density vs luminosity(from Nicer flux)")
plt.grid()
plt.xlabel("Luminosity erg/s")
plt.ylabel("x $10^{22}$ atoms/$cm^2$")
```

```python
yerr = [np.abs(df['pi_low']), np.abs(df['pi_hi'])]
plt.errorbar(df['Lumin'],df['PI'],yerr=yerr,fmt='o', capsize=5, label='PI values')
plt.title("powerlaw photon index vs luminosity(from Nicer flux)")
plt.grid()
plt.xlabel("Luminosity erg/s")
```

```python
yerr = [np.abs(df['norm_low']), np.abs(df['norm_hi'])]
plt.errorbar(df['Lumin'],df['norm'],yerr=yerr,fmt='o', capsize=5, label='norm values')
plt.title("powerlaw norm vs luminosity(from Nicer flux)")
plt.grid()
plt.xlabel("Luminosity erg/s")
```

```python
def clean_line_value(value):
    try:
        # Try converting to float
        return float(value)
    except ValueError:
        # Handle 'frozen' or other non-numeric cases
        return np.nan
df['kT'] = df['kT'].apply(clean_line_value)
df['kt_low']=df['kt_low'].apply(clean_line_value)
df['kt_hi']=df['kt_hi'].apply(clean_line_value)

yerr = [np.abs(df['kt_low']), np.abs(df['kt_hi'])]

# Plotting with error bars
plt.errorbar(df['Lumin'], df['kT'], yerr=yerr, fmt='o', capsize=5, color='teal', label='kT values')
plt.title("disk temperature vs luminosity(from Nicer flux)")
plt.grid()
plt.xlabel("Luminosity erg/s")
plt.ylabel("keV")
```
